```python
class Library:
    def __init__(self):
        self.file_name = "books.txt"
        self.file = open(self.file_name, "a+")

    def __del__(self):
        self.file.close()

    def list_books(self):
        self.file.seek(0)
        book_list = self.file.read().splitlines()
        for book in book_list:
            book_info = book.split(",")
            title = book_info[0].strip()
            author = book_info[1].strip()
            print("Title:", title, " Author:", author)


    def add_book(self):
        title = input("Enter the book title: ").strip()
        title = title.title()
        author = input("Enter the author: ").strip()
        author = author.title()
        release_year = int(input("Enter the release year: "))
        number_of_pages = int(input("Enter the number of pages: "))

        book_info = (f"{title}, {author}, {release_year}, {number_of_pages}\n")
        self.file.write(book_info)
        print()
        print(title, "added successfully.")

        self.file.seek(0)
        self.file.read()


    def remove_book(self):
      book_title = input("Enter the title of the book you want to remove: ").lower().strip()

      self.file.seek(0)
      books = self.file.readlines()
      updated_books = [book for book in books if not book.lower().startswith(book_title.lower()+ ',')]

      if len(updated_books) == len(books):
        print()
        print(f"Book {book_title.title()} not found.")
      else:
        self.file.seek(0)
        self.file.truncate()
        self.file.writelines(updated_books)
        print()
        print(f"Book {book_title.title()} removed successfully.")

        self.file.seek(0)



lib = Library()

while True:
    print()
    print("*** MENU ***")
    print("1) List Books")
    print("2) Add Book")
    print("3) Remove Book")
    print("q) Quit")

    choice = input("Enter your choice: ")

    if choice == '1':
        lib.list_books()
    elif choice == '2':
        lib.add_book()
    elif choice == '3':
        lib.remove_book()
    elif choice.lower() == 'q':
        break
    else:
        print("Invalid choice. Please enter 1, 2, 3, or q.")
```
